```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = np.array([1930,1940,1950,1960,1970,1980,1990,2000,2010])
GDP = np.array([1.015,1.33,2.29,3.26,4.951,6.759,9.366,13.131,15.599])
plt.plot(years,GDP,'ro--')
plt.xlabel('Year')
plt.ylabel('Trillion of $')
plt.title('Real US GDP (in Trillions)')
plt.show()
```

```python
np.corrcoef(years,GDP)[0,1]
```

The correlation coefficient is calculated to be .9643, which indicates that the years and the GDP are highly correlated. Since the GDP seems to be increasing with the years, there may be a positive exponential relationship between the GDP and the years. 

```python
years = np.array([1930,1940,1950,1960,1970,1980,1990,2000,2010])
GDP = np.array([1.015,1.33,2.29,3.26,4.951,6.759,9.366,13.131,15.599])
plt.plot(years,GDP,'ro--')

lowerLimit = np.log(1.015) 
upperLimit = np.log(15.599)
expX = np.linspace(lowerLimit,upperLimit,9)
plt.plot(years, np.exp(expX),'b-')

axes = plt.gca()
axes.set_ylim([0,16])
plt.xlabel('Year')
plt.ylabel('Trillion of $')
plt.title('Real US GDP (in Trillions)')
plt.show()
```
